# **Linear Regression**

**import libraries**

```python
import pandas as pd #panda for data manipulation
import numpy as np #numerical operation
import matplotlib.pyplot as plt #data visualization
from sklearn.linear_model import LinearRegression #Data Model
```

dataset

```python
data = pd.read_csv("supermarket_sales.csv") #data is variable only
```

```python
plt.scatter(data['gross income'], data['Rating'])
plt.xlabel('gross income (X)')
plt.ylabel('Rating (Y)')
plt.title('Scatter Plot of supermarket sales')
plt.show()
```

Data modelling

```python
x = data[['Rating']] #independent variable
y = data[['gross income']] #dependent variable
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
slope = model.coef_[0]

intercept = model.intercept_

print('slope: ', slope)
print('intercept:  ', intercept)
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), color='green')
plt.xlabel('Gross income (X)')
plt.ylabel(' Rating (Y)')
plt.title('Scatter Plot of Supermarket Sales')
plt.show()
```
