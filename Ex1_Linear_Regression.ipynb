# Linear Regression

# **1. Import Libraries**

```python
import pandas as pd #panda for data manipulation
import numpy as np #numerical operation
import matplotlib.pyplot as plt #data visualization
from sklearn.linear_model import LinearRegression #Data Model
```

# 2. **DataSet**


```python
x =[3, 1, 2, 6, 4]
y= [90, 85, 88, 98, 91]
```

# **3. Creating DataFrame and Saving to CSV**

```python
data = pd.DataFrame({'Study Hours': x, 'Grade': y})
data.to_csv('data.csv', index=False)
print(data.head()) #data.tail for bottom and ok lang maglagay sa loob para mabilang kung ilan lang
```

# **4. Visualizing the Data**

```python
plt.scatter(data['Study Hours'], data['Grade'])
plt.xlabel('Study Hours (X)')
plt.ylabel('Grade (Y)')
plt.title('Scatter Plot of Student Performance')
plt.show()
```

# **5. Preparing Data for Modelling**

```python
x = data[['Study Hours']] #independent variable
y = data[['Grade']] #dependent variable
```

# **6. Fitting the Linear Regression Model**

```python
model = LinearRegression()
model.fit(x,y)
```

# **7. Getting Model Coefficients**

```python
slope = model.coef_[0]

intercept = model.intercept_

print('slope: ', slope)
print('intercept:  ', intercept)
```

# **8. Visualizing the Linear Regression Line**

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), color='green')
plt.xlabel('Study Hours (X)')
plt.ylabel('Grade (Y)')
plt.title('Scatter Plot of Student Performance')
plt.show()
```

# **9. Making Predictions**

```python
input_value = np.array([[0.5]])
predicted_y = model.predict(input_value)
print(predicted_y)
```
